```python
import pandas as pd

housing_file_path = '../content/housing.csv'
housing_data = pd.read_csv(housing_file_path)
housing_data.columns
```

```python
housing_data = housing_data.dropna(axis=0)
```

```python
y = housing_data.median_house_value
```

```python
housing_features = ['population', 'households', 'total_rooms', 'median_income', 'latitude', 'longitude']
```

```python
X = housing_data[housing_features]
```

```python
from sklearn.tree import DecisionTreeRegressor

# Define model. Specify a number for random_state to ensure same results each run
housing_model = DecisionTreeRegressor(random_state=1)

# Fit model
housing_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head())
print("The predictions are")
print(housing_model.predict(X.head()))
```
